Before we dive into the analysis, we need to first define some objective metrics for measuring the quality of a piece of work. 

1. The first variable to use will be text complexity, which refers to the number of unique words per length of the text.
2. Second, we will use a metric measuring the diversity of the text (the proportion of all parts of speech)
3. Third, proportion of complex words (words longer than 7 letters)
4. Fourth, proportion of complex sentences (sentences longer than 7 words)
5. Fifth, language level analysis

```python
#lemmatize words
unique_words = []
complex_words = []
easy_words = []
grade_level = []
readibility = []
sentence_comp = []
s_1 = []
s_8 = []

def get_metrics(season):
    print(season)
    df_s = df[df.Season == "Season "+ str(season)]
    list_sents = []
    if season < 7:
        for i in range(1,11):
            df_s_e = df_s[df_s.Episode == "Episode "+ str(i)]
            list_sents.append(str(list(df_s_e.Sentence)))
    if season == 7:
        for i in range(1,8):
            df_s_e = df_s[df_s.Episode == "Episode "+ str(i)]
            list_sents.append(str(list(df_s_e.Sentence)))
    if season == 8:   
        for i in range(1,7):
            df_s_e = df_s[df_s.Episode == "Episode "+ str(i)]
            list_sents.append(str(list(df_s_e.Sentence)))
    
    
    #unique words
    tknzr = TweetTokenizer()
    l = []
    for i in list_sents: 
        y = tknzr.tokenize(i)
        new_y = []
        for w in y:
            w = ps.stem(w)
            w = lemmatizer.lemmatize(w)
            new_y.append(w)   
        unique_w = set(list(new_y))
        l.append(len(unique_w)/len(new_y)*100)
    print("Unique Words", l)    
    unique_words.append(np.mean(l))    
    
    #complex words
    tknzr = TweetTokenizer()
    s = []
    count = 0
    for i in list_sents: 
        y = tknzr.tokenize(i)
        new_y = []
        new = []
        for w in y:
            w = ps.stem(w)
            w = lemmatizer.lemmatize(w)
            new.append(w)
            if len(w) >= 7:
                count += 1
                new_y.append(w) 
        s.append(len(new_y)/len(new)*100)
    print("Complex Words", s)     
    complex_words.append(np.mean(s))   
    
    #easy_words
    tknzr = TweetTokenizer()
    d = []
    count = 0
    for i in list_sents: 
        y = tknzr.tokenize(i)
        new_y = []
        new = []
        for w in y:
            w = ps.stem(w)
            w = lemmatizer.lemmatize(w)
            new.append(w)
            if len(w) <= 5:
                count += 1
                new_y.append(w) 
        d.append(len(new_y)/len(new)*100)
    print("Easy Words", d)     
    easy_words.append(np.mean(d))  
    
    #grade_level and readibility
    grade_level.append(textstat.flesch_reading_ease(str(list_sents)))
    readibility.append(textstat.flesch_kincaid_grade(str(list_sents)))

    
    #sentence_complexity
    r = []
    for i in list_sents:
        y = sent_tokenize(i)
        counts = 0
        for e in y:
            f = tknzr.tokenize(e)
            if len(f) >= 15:
                counts += 1
        z = counts/len(y)*100        
        r.append(z)  
    print("Unique Words", r)     
    sentence_comp.append(np.mean(r))
    
    
    
    
for i in range(1,9):
    get_metrics(i)
```
